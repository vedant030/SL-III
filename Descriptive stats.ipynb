```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=({
    "Age":np.random.choice(['Young','Adult','Mid-Age','Old'],50),
    "Income":np.random.randint(20000,50000,50)
})
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
summary_stats=df.groupby('Age')['Income'].agg(['mean','median','min','max','std'])
```
